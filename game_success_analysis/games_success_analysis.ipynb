# Исследование популярных игр

К нам обратился разработчик компьютерных игр с целью выявить определяющие успешность игры закономерности. Это позволит сделать ставку на потенциально популярный продукт и спланировать разработку более популярной игры, а также ее будущие рекламные кампании.  
   
Для этого была предоставлена выборка по историческим данным о продажах игр, оценках пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation). Предоставлены данные до 2016 года. Также была предоставлена документация с описанием данных. 
    
**Описание данных в таблице `games`:**  

* `Name` — название игры
* `Platform` — платформа
* `Year_of_Release` — год выпуска
* `Genre` — жанр игры
* `NA_sales` — продажи в Северной Америке (миллионы проданных копий)
* `EU_sales` — продажи в Европе (миллионы проданных копий)
* `JP_sales` — продажи в Японии (миллионы проданных копий)
* `Other_sales` — продажи в других странах (миллионы проданных копий)
* `Critic_Score` — оценка критиков (максимум 100)
* `User_Score` — оценка пользователей (максимум 10)
* `Rating` — рейтинг от организации ESRB (англ. Entertainment Software Rating Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию.
    
 > Примечание: Данные за 2016 год могут быть неполными.

**Цель исследования** — выявить определяющие успешность игры закономерности.

## Подготовка данных

### Изменение названий столбцов

Необходимо привести названия столбцов к нижнему регистру.

```python
# Приводим названия к нижнему регистру
games.columns = games.columns.str.lower()
# Проверяем
games.columns
```

Названия столбцов приведены к нижнему регистру.

### Работа с пропусками

Необходимо проанализировать пропуски в столбцах и решить - оставить их либо заполнить.

```python
# Посчитаем, сколько пропусков в каких столбцах.
print(games.isna().sum())
# Посчитаем, сколько пропусков в процентах.
games.isna().mean()
```

В столбцах с названием игры, жанром и годом выпуска пропуски занимают менее 2 процентов - эти пропуски удалим, т.к. их нет возможности заполнить валидными данными.

```python
# Удаляем строки, где есть пропуски в этих столбцах.
games.dropna(subset=['name', 'year_of_release', 'genre'], inplace=True)
# Проверяем удаление.
games.isna().sum()
```

В остальных столбцах с пропусками слишком много данных, чтобы их удалять. Заполним пропуски в столбцах с оценкой значением "-1", а в столбце с рейтингом "unknown", чтобы не смешивать типы данных и чтобы бывшие пропуски можно было потом отфильтровать.

```python
# Заполняем пропуски в рейтинге
games['rating'].fillna('unknown', inplace=True)

# Заполняем пропуски в оценках (оставшиеся столбцы).
games.fillna(-1, inplace=True)

# Проверяем.
games.isna().sum()
```

### Приведение типов

Неоходимо в столбце года привести тип к int как более соответствующему, в столбце пользовательских оценок - к float. Проверим типы остальных столбцов.

```python
# Посмотрим типы
games.dtypes
```

```python
# Приводим тип года к int
games['year_of_release'] = games['year_of_release'].astype('int')
# Проверяем
games.dtypes
```

Посмотрим, какие данные есть в столбце пользовательских оценок.

```python
games['user_score'].unique()
```

Видно, что все числа являются строками, также интерес представляет значение `tbd`. Поиск в интернете раскрывает, что так обозначаются значения, которые только будут определены. По сути, это пропуски. Возможно, недостаточно оценок, чтобы рассчитать оценку. Эти значения заменим также как и пропуски на "-1".

```python
# Заменим тип на float, заменяя значения, которые нельзя привести на NaN
games['user_score'] = pd.to_numeric(games['user_score'], errors='coerce')
# Заменим пропуски на -1
games = games.fillna(-1)
# Проверим
print(games.dtypes)
games.isna().sum()
```

Типы данных в столбцах года выпуска и пользовательских оценок заменены на соответствующие. В остальных столбцах в типах нет ничего необычного.

## Исследовательский анализ данных

### Исследование продаж игр

Посмотрим, сколько игр выпускалось в разные годы.

```python
# Установим размеры всех графиков 15 на 10 дюймов.
sns.set(rc={'figure.figsize':(15, 10)})
```

```python
count_games = games.groupby('year_of_release', as_index=False)['name'].count()

# Строим столбчатую по количеству выпущенных игр за каждый год.
sns.barplot(data=count_games, x='year_of_release', y='name')

# Повернем подписи на 45 градусов по х
plt.xticks(rotation=45)

plt.title('Количество выпущенных игр в каждый год')
plt.xlabel('Год выпуска')
plt.ylabel('Количество игр')

# Добавим медиану на график
median = stat.median(count_games['name'])
plt.axhline(median, linestyle='dashed')

plt.show()
```

По графику видно, что в каждый год, начиная с 1998 выпускалось игр более медианного значения. При этом явный всплеск количества игр за год виден в 2008-2009 годах.

Посмотрим, как менялись продажи по платформам. Выберем топ-5 платформ с наибольшими суммарными продажами и построим распределение продаж их игр по годам.

По верхнему значению можно заметить, что тут что-то не так, т.к. платформа DS была выпущена только в 2004 году, согласно данным из интернета. Проверим данные по этой платформе подробнее. 

```python
games_top.query('(platform == "DS") & (year_of_release < 2004)')
```

Какая-то ошибка в данных. Для этой платформы есть одна игра, которая "выпущена" ранее года выпуска консоли. Судя по интернету, данная игра выпущена в 2010 году. Исправим ошибку в изначальном сете, затем заново сделаем срез.

```python
games.loc[15957, 'year_of_release'] = 2010
games_top = games.query('platform in @top5')
games_top.groupby(['year_of_release', 'platform'], as_index=False)['total_sales'].sum().head()
```

Ошибка исправлена, можно строить график.

```python
games_top_by_year = games_top.groupby(['year_of_release', 'platform'], as_index=False)['total_sales'].sum()
ax = sns.lineplot(data=games_top_by_year, x='year_of_release', y='total_sales', hue='platform')

ax.set_xticks(np.arange(2000, 2017, 1))

# Повернем подписи на 45 градусов по х
plt.xticks(rotation=45)

plt.title('Продажи игр по годам на топ-5 платформах')
plt.xlabel('Год выпуска')
plt.ylabel('Количество продаж')

plt.legend(loc='upper left')
plt.show()
```

По графику видно, что срок жизни всех топ-платформ 8-10 лет, с резким подъемом в течение двух лет, периодом пика примерно с 3го по 6й год, и далее период отмирания с 7го по 10й год.

Выберем актуальный период для дальнейшего исследования. Так как мы хотим сделать прогноз на 2017 год, то необходимо выбрать период до 2016 (включительно), а т.к. период до пика у платформ примерно 5 лет, то длину интервала выберем в 5 лет. Таким образом, интересующий нас актуальный период будет с 2012 по 2016 год.

```python
# Сделаем срез по актуальной дате
years = [x for x in range(2012,2017)]
games_actual = games.query('year_of_release in @years').copy()
games_actual.head()
```

### Выбор потенциально прибыльных платформ

Продолжим исследование в актуальном периоде.  
Ответим, какие платформы лидируют по продажам, растут или падают? Выберим несколько потенциально прибыльных платформ.  
Для этого построим график всех продаж по платформам за актуальный период.

```python
games_actual_by_year = games_actual.groupby(['platform', 'year_of_release'], as_index=False)['total_sales'].sum()
ax = sns.lineplot(data=games_actual_by_year, x='year_of_release', y='total_sales', hue='platform')

ax.set_xticks(np.arange(2012, 2017, 1))

plt.title('Продажи игр в актуальном периоде')
plt.xlabel('Год выпуска')
plt.ylabel('Количество продаж')

plt.legend(loc='upper left')
plt.show()
```

```python
sales_actual = games_actual.groupby('platform')['total_sales'].sum().sort_values(ascending=False).reset_index()[0:5]
# Обновим топ-5 для актуального периода
top5 = sales_actual['platform']
sales_actual
```

По представленным данным видно, что наибольшие продажи в актуальном периоде на доле платформ PS4, PS3, X360, 3DS и XOne, однако, они уже снижают свою активность к концу выбранного актуального периода. На всех платформах продажи снижаются.

### Влияние отзывов на продажи

Посмотрим, как влияют на продажи внутри одной популярной платформы (по списку топ-5 платформ по продажам) отзывы пользователей и критиков. Построим диаграммы рассеяния и посчитаем корреляцию между отзывами и продажами.  
Помним, что в оценках были пропуски, которые мы заменили на -1, поэтому срезы будем делать с учетом этого.

```python
# Построим диаграммы рассеяния по оценкам критиков и пользователей по топ-5 платформам
for plat in top5:
    # Делаем срез для оценок критиков
    games_platform_critics = games_actual.query('(platform == @plat) & (critic_score >= 0)').copy()
      
    # Делаем срез для оценок пользователей
    games_platform_users = games_actual.query('(platform == @plat) & (user_score >= 0)').copy()
    
    # Строим график для оценок критиков
    critic_plot = sns.scatterplot(data=games_platform_critics, 
                                  x='critic_score', 
                                  y='total_sales', 
                                  )
    plt.title(f'Диаграмма рассеяния оценок критиков для платформы {plat}')
    plt.ylabel('Количество продаж')
    plt.xlabel('Оценка')
    plt.show()
    
    # Строим график для оценок пользователей
    user_plot = sns.scatterplot(data=games_platform_users, 
                                x='user_score', 
                                y='total_sales', 
                                )
    plt.title(f'Диаграмма рассеяния оценок пользователей для платформы {plat}')
    plt.ylabel('Количество продаж')
    plt.xlabel('Оценка')
    plt.show()
```

По диаграммам можно сказать, что по всем топ-платформам количество продаж имело слабую зависимость от оценок критиков и пользователей, т.к. по большинству игр (которые проданы тиражами до 2 млн копий) оценки строятся почти равномерно. Однако, можно заметить, что большое количество продаж (выбросы более 10 млн продаж) у игр с большой оценкой (более 7 у пользователей, более 70 у критиков).

```python
# Построим таблицы корреляции по оценкам критиков и пользователей по топ-5 платформам
for plat in top5:
    # Делаем срез для оценок критиков
    games_platform_critics = games_actual.query('(platform == @plat) & (critic_score >= 0)').copy()
    print(f'Коэффициент корреляции между оценками критиков и продажами платформы {plat}:', 
         games_platform_critics['total_sales'].corr(games_platform_critics['critic_score']))

    # Делаем срез для оценок пользователей
    games_platform_users = games_actual.query('(platform == @plat) & (user_score >= 0)').copy()
    print(f'Коэффициент корреляции между оценками пользователей и продажами платформы {plat}:', 
         games_platform_users['total_sales'].corr(games_platform_users['user_score']))
    
    print('-'*100)
```

## Проверка гипотез

### О равенстве средних пользовательских рейтингов платформ Xbox One и PC

Необходимо проверить, что средняя оценка пользователей платформ Xbox One и PC различаются. Сформулирем нулевую и альтернативную гипотезы:
* H0 - средняя оценка пользователей платформ Xbox One и PC равны;
* H1 - средняя оценка пользователей платформ Xbox One и PC не равны.

Разобьем таблицу `games` на две фильтрованные таблицы - по каждой платформе с данными об оценках пользователей.  
Далее сделаем ttest на проверку нулевой гипотезы с доверительным интервалом 5%.

```python
# Создаем сеты с данными об оценках пользователей двух платформ.
xone = games.loc[(games['platform'] == 'XOne') & (games['user_score'] >= 0)]['user_score']
pc = games.loc[(games['platform'] == 'PC') & (games['user_score'] >= 0)]['user_score']

# Проверяем гипотезу о равенстве среднего. 
alpha = 0.05
results = st.ttest_ind(xone, pc)
print('p-значение: ', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвернуть нулевую гипотезу')
```

Нулевая гипотеза о том, что средняя оценка пользователей платформ XboxOne и PC равна не подтвердилась.  
Принимаем альтернативную гипотезу.

**Вывод**: средняя оценка пользователей платформ XboxOne и PC не равна.

Выясним, какая оценка больше. Проведем односторонний ttest.  
Нулевая гипотеза - средняя оценка пользователей платформ Xbox One и PC равны.  
Альтернативная гипотеза - средняя оценка пользователей платформы XboxOne меньше оценки пользователей на PC.

```python
results = st.ttest_1samp(xone, pc.mean(), alternative='less')
print('p-значение: ', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу, оценки на Xbox значимо меньше оценок на PC")
else:
    print("Не отвергаем нулевую гипотезу, оценки пользователей Xbox достигают оценок на PC") 
    
print('Средняя оценка пользователей игр XboxOne по выборке:', xone.mean())
print('Средняя оценка пользователей игр ПК по выборке:', pc.mean())
```

По однострононнему ttest можно сделать вывод, что оценки пользователей на Xbox значимо меньше оценок на PC.  
Следовательно, оценки на PC значимо больше оценок на XboxOne - по среднему видно, что примерно на полбалла.

### О равенстве средних пользовательских рейтингов жанров Action и Sports

Необходимо проверить, что средняя оценка пользователей жанров Action и Sports различаются. Сформулирем нулевую и альтернативную гипотезы:
* H0 - средняя оценка пользователей жанров Action и Sports равны;
* H1 - средняя оценка пользователей жанров Action и Sports не равны.

Разобьем таблицу `games` на две фильтрованные таблицы - по каждому жанру с данными об оценках пользователей.  
Далее сделаем ttest на проверку нулевой гипотезы с доверительным интервалом 5%.

```python
# Создаем сеты с данными об оценках пользователей двух жанров.
action = games.loc[(games['genre'] == 'Action') & (games['user_score'] >= 0)]['user_score']
sports = games.loc[(games['genre'] == 'Sports') & (games['user_score'] >= 0)]['user_score']

# Проверяем гипотезу о равенстве среднего. 
alpha = 0.05
results = st.ttest_ind(action, sports)
print('p-значение: ', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвернуть нулевую гипотезу')
    
print('Средняя оценка пользователей игр жанра Action по выборке:', action.mean())
print('Средняя оценка пользователей игр жанра Sports по выборке:', sports.mean())
```

Нулевая гипотеза о том, что средняя оценка пользователей жанров Action и Sports равна подтвердилась.  
Принимаем нулевую гипотезу.

**Вывод**: средняя оценка пользователей жанров Action и Sports статистически значимо равны (около 7 баллов).
