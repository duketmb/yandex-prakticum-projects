## Предобработка данных

В обзоре данных были определены проблемы, которые необходимо решить. Рассмотрим каждый "проблемный" столбец отдельно подробнее.  
Начнем с `locality_name` - проверим на пропуски, исправим "ё".

### Обработка названий городов

Необходимо в стоблце `locality_name` проверить пропуски и привести стиль написания всех слов через "е".


```python
# Общая информация о столбце.
estate['locality_name'].describe()
```

Как видно из общего описания - в столбце присутствуют пропуски, т.к. всего в таблице 23699 строк, а в исследуемом столбце непустых - 23650. Также из общего описания интересно, что уникальных значений - 364 (и к ним добавятся 365й -  неизвестный город), а самый популярный - Санкт-Петербург.  
Заполним пропуски названием города "неизвестный".

```python
# Заполняем пропуски.
estate['locality_name'] = estate['locality_name'].fillna('неизвестный')
# Проверяем наличие пропусков.
estate['locality_name'].isna().sum()
```

Заменим стиль написания буквы ё на "е" для единообразия, а также для исключения дубликатов. Также для исключения дубликатов, приведем все значения к нижнему регистру.

```python
# Заменяем `посёлки` на `поселки`.
estate['locality_name'] = estate['locality_name'].str.replace('ё', 'е')
# Приводим к нижнему регистру.
estate['locality_name'] = estate['locality_name'].str.lower()
# Смотрим, сколько уникальных значений осталось.
len(estate['locality_name'].unique())
```

Осталось 331 уникальное значение имени города (а было 365), значит, дубликаты среди городов все-таки были.

### Обработка расстояния до центра

Перейдем к `cityCenters_nearest` - переименуем, заполним пропуски и переведем значения в километры.  


```python
# Общая информация о столбце.
estate['cityCenters_nearest'].describe()
```

```python
# Переименовываем название столбца.
estate = estate.rename(columns={'cityCenters_nearest': 'centers_nearest'})
# Проверяем, что столбец переименован.
estate.columns
```

Для того, чтобы понять, как заполнить пропуски в этом столбце, построим сводную таблицу по значениям этого столбца, сгруппированные по городам.

```python
# Сводная таблица по городам, в значениях - количество, среднее и медиана удаленности от центра.
pivot_cities = estate.pivot_table(
    index='locality_name',
    values='centers_nearest',
    aggfunc=['count', 'mean', 'median']
)
# Убираем уровень мультииндекса.
pivot_cities = pivot_cities.droplevel(1, axis=1).sort_values('count')
# Проверяем таблицу.
pivot_cities
```

```python
# Посмотрим на города, в которых частично пропущены значения.
pivot_cities[pivot_cities['count'] != 0]
```

Видно, что в некоторых городах нет ни одного объявления, в котором указано расстояние до центра, а в некоторых частично пропущены значения.  
Для городов с частично пропущенными значениями заполним пропуски средним расстоянием до центра по этому городу, т.к. из таблицы выше видно, что разница между средним и медианой небольшая, а значит, можно предположить, что аномальных значений нет (в среднем из одного и того же города до центра примерно одинаковое расстояние).  
Для городов с полностью пропущенными значениями заполним максимальным значением из всего столбца, т.к. неизвестно, сколько добираться до центра, но скорее всего в среднем не больше, чем максимум по столбцу.  


```python
# Присвоим переменной maximum максимальное значение по столбцу.
maximum = estate['centers_nearest'].max()
# Для каждого уникального города из фрейма, вычислим среднее по столбцу и заполним им пропуски.
# Для городов с полностью пропущенными значениями среднее не вычислится и пропуски останутся.
# Оставшиеся пропуски заполним значением переменной maximum.
for city in estate['locality_name'].unique():
    mean = estate.loc[estate['locality_name']
                        == city, 'centers_nearest'].mean()
    estate.loc[estate['locality_name'] 
               == city, 'centers_nearest'] = estate.loc[estate['locality_name'] 
                                                        == city, 'centers_nearest'].fillna(mean)
estate['centers_nearest'].fillna(maximum, inplace=True)
# Проверим, что пропусков не осталось.
estate['centers_nearest'].isna().sum()
```

Приведем все значения столбца к километрам для удобства, округляя до 2 знаков после запятой.

```python
# Каждое значение делим на 1000 и округляем.
estate['centers_nearest'] = round(estate['centers_nearest'] / 1000, 2)
# Проверяем округление.
estate['centers_nearest']
```

Столбец переименован, в нем заполнены пропуски и переведены значения в километры.

### Обработка количества этажей

Проверим столбец на пропуски и изменим тип значений, т.к. количество этажей не может быть не целым.

```python
# Количество пропусков в столбце floors_total.
estate['floors_total'].isna().sum()
```

Всего 86 значений пропущено в данном столбце. Отсутствовать они могут по следующим причинам:
* этаж всего один и не было смысла указывать сколько всего этажей;
* этаж не указан, потому что квартира на последнем этаже, а данные об этом потеряны из выгрузки;
* этаж не указан по ошибке.

В любом случае, из почти 24 тысяч значений пропущено всего 86. Заменим пропуски на значения из столбца `floor` той же строки (будто указанный этаж - последний).

```python
# Присваиваем пропускам значения из столбца floor.
estate.loc[estate['floors_total'].isna(), 
           'floors_total'] = estate.loc[estate['floors_total'].isna(), 'floor']
# Проверяем количество пропусков.
estate['floors_total'].isna().sum()
```

В столбце `floors_total` пропущенные значения заменены на значения столбца `floor` в аналогичных строках. 

### Обработка высоты потолков

В столбце `ceiling_height` необходимо заполнить пропуски.  
Посмотрим на общую информацию о столбце.

```python
estate['ceiling_height'].describe()
```

Видно, что среди данных есть аномалии (потолки с высотой 100 метров). Поэтому пропуски правильнее заполнить медианой. Однако, застройка в основном типовая и в домах с одинаковым количеством этажей потолки примерно одной высоты.  
Значит, заполним пропуски медианной высотой для каждого дома с одинаковой этажностью. Для некоторой этажности могут быть только пропущенные значения - такие заполним медианой по всему столбцу.

```python
# Заполняем пропуски медианой по домам аналогичной этажности.
estate['ceiling_height'].fillna(estate.groupby('floors_total')['ceiling_height'].transform('median'), inplace=True)
# Для оставшихся пропусков высоту заполним медианой по всему столбцу.
estate['ceiling_height'].fillna(estate['ceiling_height'].median(), inplace=True)
# Проверим, что пропусков не осталось.
estate['ceiling_height'].isna().sum()
```

### Обработка жилой площади и площади кухни


В столбцах `living_area` и `kitchen_area` необходимо заполнить пропуски.  
Возможно, пропуски образовались из-за того, что значения очевидны (разница между имеющимися значениями), либо из-за ошибки выгрузки (т.к. есть строки, где пропущена и площадь кухни, и жилая площадь).  
Жилая площадь это разница между общей площадью и площади кухни.  
Площадь кухни это разница между общей площадью и жилой площадью.  
Поэтому заполним пропуски разницей значений соответствующих столбцов.  
Однако, таким образом не получится заполнить пропуски в строках, где одновременно пропущены значения общей площади и площади кухни.  
Добавим к фрейму два столбца:
- `kitchen_total` - отношение площади кухни к общей площади;
- `living_total` - отношение жилой площади к общей площади.

Остальные пропуски в столбцах `living_area` и `kitchen_area` заполним значениями столбца `total_area`, домноженным на медианное значение соответствующих столбцов `kitchen_total` и `living_total`.

```python
# Заполняем пропуски общей площади разницей общей площади и площади кухни.
estate['living_area'] = estate['living_area'].fillna(estate['total_area'] - estate['kitchen_area'])
# Заполняем пропуски площади кухни разницей общей площади и жилой площади.
estate['kitchen_area'] = estate['kitchen_area'].fillna(estate['total_area'] - estate['living_area'])
# Смотрим, сколько пропусков осталось.
estate.loc[:, ['living_area', 'kitchen_area']].isna().sum()
```

```python
# Считаем отношение площади кухни к общей площади.
estate['kitchen_total'] = estate['kitchen_area'] / estate['total_area']
# Заполним пропуски в столбце kitchen_area через total_area путем домножения 
# на медианное значение столбца kitchen_total
estate['kitchen_area'].fillna(estate['total_area'] * estate['kitchen_total'].median(), inplace=True)
# Считаем отношение жилой площади к общей площади.
estate['living_total'] = estate['living_area'] / estate['total_area']
# Заполним пропуски в столбце living_area через total_area путем домножения 
# на медианное значение столбца living_total
estate['living_area'].fillna(estate['total_area'] * estate['living_total'].median(), inplace=True)
# Проверяем оставшиеся пропуски
estate.loc[:, ['living_area', 'kitchen_area']].isna().sum()
```

В столбцах `living_area` и `kitchen_area` заполнены пропуски.

### Фильтрация аномальных значений

Удалим из изученных столбцов аномальные значения. Аномальные значения будем выбирать среди следующих значений по столбцам:
* `total_area`;
* `last_price`;
* `rooms`;
* `ceiling_height`;
* `days_exposition`.  

Убрать необходимо не более 10%, а т.к. параметров много, проверим, сколько значений будет убрано, если отсечь по 1% сверху и снизу с каждого параметра. При этом не забудем, что в `days_exposition` изначальные пропуски были заменены на отрицательные значения - по этому столбцу будем фильтровать только до верхней границы.

```python
# Задаем верхнюю и нижнюю границы
low_percentele = 0.01
hihg_percentele = 0.99

# Рассчитываем границы по каждому параметру
total_area_low = estate['total_area'].quantile(low_percentele)
total_area_high = estate['total_area'].quantile(hihg_percentele)
last_price_low = estate['last_price'].quantile(low_percentele)
last_price_high = estate['last_price'].quantile(hihg_percentele)
rooms_low = estate['rooms'].quantile(low_percentele)
rooms_high = estate['rooms'].quantile(hihg_percentele)
ceiling_height_low = estate['ceiling_height'].quantile(low_percentele)
ceiling_height_high = estate['ceiling_height'].quantile(hihg_percentele)
days_exposition_high = estate['days_exposition'].quantile(hihg_percentele)
# Выводим их не экран
print(f'Границы total_area: {total_area_low} - {total_area_high}')
print(f'Границы last_price: {last_price_low} - {last_price_high}')
print(f'Границы rooms: {rooms_low} - {rooms_high}')
print(f'Границы ceiling_height: {ceiling_height_low} - {ceiling_height_high}')
print(f'Границы days_exposition: до {days_exposition_high}')

# Считаем и выводим процент отбрашенных значений
print(
    'Процент отброшенных значений =',
    round(
        100 - len(estate.query('@total_area_low <= total_area <= @total_area_high\
                                    & @last_price_low <= last_price <= @last_price_high\
                                    & @rooms_low <= rooms <= @rooms_high\
                                    & @ceiling_height_low <= ceiling_height <= @ceiling_height_high\
                                    & days_exposition <= @days_exposition_high'))
        / len(estate) * 100,
        2
    ),
    '%'
)
```

При таких границах фильтрации от фрейма будет отброшено 6,3% данных (не учитывая изначальные пропуски по столбцу `days_exposition`). Значит, можно отбросить эти данные по таким границам.

```python
# Фильтруем данные в новую таблицу estate_good
estate_good = estate.query('@total_area_low <= total_area <= @total_area_high\
                                    & @last_price_low <= last_price <= @last_price_high\
                                    & @rooms_low <= rooms <= @rooms_high\
                                    & @ceiling_height_low <= ceiling_height <= @ceiling_height_high\
                                    & days_exposition <= @days_exposition_high')
# Проверяем процент остатка данных
len(estate_good)/len(estate)
```

Отфильтрованных данных осталось 93,7% от изначальных.

### Определение населенных пунктов с дорогими и дешевыми квартирами

Определим, в каких населенных пунктах самые дорогие и самые дешевые по стоимости квадратного метра квартиры.  
Для этого:
* выберем 10 населенных пунктов с наибольшим числом объявлений;
* рассчитаем среднюю цену квадратного метра в каждом из них;
* отсортируем по средней цене.

```python
# Создаем список городов с наибольшим числом объявлений top_cities.
top_cities = (
              # Выбираем из "хорошей" таблицы.
              estate_good
              # Группируем по названию населенного пункта.
              .groupby('locality_name')
              # Считаем число объявлений.
              .count()
              # Сортируем по столбцу без пропусков (rooms), в порядке убывания.
              .sort_values(by='rooms', ascending=False)
              # Берем топ-10 результатов
              .head(10)
              # из индексов (индексы-названия населенных пунктов в GroupBy)
              .index
              )
# Проверяем список.
top_cities
```

```python
# Считаем среднюю цену метра по городам с наибольшим количеством объявлений.
top_10_cities = (
    # Создаем срез из "хорошего" фрейма по городам с наибольшим количеством объявлений.
    estate_good.query('locality_name in @top_cities')
    # Группируем по названию населенного пункта, выбирая столбец с ценой квадратного метра.
    .groupby('locality_name')['price_sqm']
    # Рассчитываем среднее.
    .mean()
    # Сортируем по убыванию.
    .sort_values(ascending=False)
    # Сбросим индекс, чтобы был полноценный фрейм.
    .reset_index()
)
top_10_cities
```

```python
# Построим столбчатую диаграмму цены на квадратный метр по самым популярным городам.
top_10_cities.plot(kind='bar',
                   x='locality_name',
                   xlabel='Населенный пункт',
                   y='price_sqm',
                   ylabel='Средняя цена квадратного метра',
                   title='Средняя цена квадратного метра по населенным пунктам',
                   grid=True,
                   figsize=(10,10),
                   legend=False,
                   rot=45
                  );
```

Наибольшая цена квадратного метра в Санкт-Петербурге - 110 тысяч. Самые дешевый квадратный метр в Выборге - 58361.

    Типичная квартира в центре площадью от 60 до 100 квадратных метров с 2-3 комнатами, ценой около 10 миллионов и высотой потолков от 2,7 до 3,2 метров.

7. Стоимость всех квартир, включая квартиры в центре, прямо зависят от одинаковых параметров: площадь и количество комнат. Также есть закономерность, что квартиры не на первом этаже стоят дороже.
